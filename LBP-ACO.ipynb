```python
import cv2
import numpy as np
from skimage import feature
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import time
from ACO_LBP_Positions_Source_Code.featureselector import FeatureSelector
```

```python
# load dataset
with open('X-Train.npy', 'rb') as f:
    X_train = np.load(f)
    
with open('Y-Train.npy', 'rb') as f:
    Y_train = np.load(f)
    
with open('X-Test.npy', 'rb') as f:
    X_test = np.load(f)
    
with open('Y-Test.npy', 'rb') as f:
    Y_test = np.load(f)
    
```

```python
if __name__ == "__main__":

    # TIME START
    start_time = time.time()


    X_train = X_train
    y_train = Y_train
    X_test  = X_test
    y_test  = Y_test



    # ACO CALL
    number_ants = 30
    number_elitist_ants = 3
    iter = 5
    version = "RACOFS"
    
    final_subset = []
    optimizer = FeatureSelector(version, X_train, y_train, X_test, y_test, number_ants, number_elitist_ants, iter, 1, 1, 0.5, 0.2, 0.08)
    final_subset = optimizer.acoFS()
    print(final_subset)

    # TIME STOP
    stop_time = time.time()
    elapsed_time = stop_time - start_time

    # OUTPUT
    print("Elapsed time:")
    print("\t", elapsed_time, "seconds")
   
```

```python
X_train_subset = X_train[:,final_subset]
X_test_subset = X_test[:,final_subset]
knn = KNeighborsClassifier(n_neighbors=3,metric="cosine")
knn.fit(X_train_subset,y_train)


pos = []
for i in range(256):
    if i in final_subset:
        pos.append(1)
    else:
        pos.append(0)
pos = np.array(pos)
print(pos,sum(pos),len(final_subset),knn.score(X_test_subset,y_test))


with open('ACO-LBP-Positions(1).npy', 'wb') as f:
    np.save(f, pos)
```
